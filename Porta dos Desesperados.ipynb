```python
from matplotlib import pyplot as plt
from random import randint
import random

quantidadeVitoriasSemTrocar = 0
quantidadeVitoriasTrocando = 0
experimento = 1
eixoX = []
eixoYSemTrocar = []
eixoYTrocando = []

while (experimento <= 1000):
    portasQueSobraram = []
    portaQueContemPremio = randint(1, 3)
    portaEscolhida = randint(1, 3)

    portasQueSobraram.append(portaEscolhida)

    if portaEscolhida == portaQueContemPremio:
      if portaQueContemPremio == 3:
        portasQueSobraram.append(1)
      else:
        portasQueSobraram.append(portaEscolhida+1)
    else:
        portasQueSobraram.append(portaQueContemPremio)

    # Estratégia sem trocar de porta
    if portaEscolhida == portaQueContemPremio:
        quantidadeVitoriasSemTrocar += 1

    # Estratégia ao trocar de porta
    novaPortaEscolhida = portasQueSobraram[1]
    if novaPortaEscolhida == portaQueContemPremio:
        quantidadeVitoriasTrocando += 1

    experimento += 1
    proporcaoSemTrocar = quantidadeVitoriasSemTrocar/experimento
    proporcaoTrocando = quantidadeVitoriasTrocando/experimento
    eixoX.append(experimento)
    eixoYSemTrocar.append(proporcaoSemTrocar)
    eixoYTrocando.append(proporcaoTrocando)

plt.plot(eixoX, eixoYSemTrocar)
plt.plot(eixoX, eixoYTrocando)

plt.title('Porta dos Desesperados')
plt.ylabel('Proporção acumulada de vitórias')
plt.xlabel('Número do Experimento')
plt.show()
```
